### 2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.

```python
import pandas as pd
page = requests.get("https://www.imdb.com/list/ls091520106/")
page
soup = BeautifulSoup(page.content)
soup

movies = []
for main_file in soup.find_all('div',class_="lister-item mode-detail"):
    nmae = main_file.h3.a.text
    nmae
    rating = main_file.find('span',class_="ipl-rating-star__rating").text
    rating
    yor = main_file.find('span',class_="lister-item-year text-muted unbold").text
    yor
    movies.append([nmae,rating,yor])
movies

df = pd.DataFrame(movies,columns=['Name','Ratings','Yor'])
df
```

# 9) Write a python program to scrape mentioned details from dineout.co.in :
i) Restaurant name

ii) Cuisine

iii) Location

iv) Ratings

v) Image URL

```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get("https://www.dineout.co.in/delhi-restaurants/buffet-special")
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
import pandas as pd
title = []

for main_file in soup.find_all('div',class_="restnt-card restaurant"):
    
    restnt_name = main_file.find('a',class_="restnt-name ellipsis").text
    cuisine = main_file.find('span',class_="double-line-ellipsis").a.text
    loc = main_file.find('div',class_="restnt-loc ellipsis").a.text
    ratings = main_file.find('div',class_="restnt-rating rating-4").text
    title.append([restnt_name,cuisine,loc,ratings])  
    
l = []
for img in soup.findAll('img',class_="no-img"):
    l.append(img['data-src'])
df = pd.DataFrame(title,columns=['restaurant name','cuisine','location','ratings'])
df1 = pd.DataFrame(l)
df['img_url'] = df1
df

```

### 10) Write a python program to scrape first 10 product details which include product name , price , Image URL from
https://www.bewakoof.com/women-tshirts?ga_q=tshirts .

```python
import selenium
from bs4 import BeautifulSoup

from selenium import webdriver
PATH= "C:\Program Files (x86)\chromedriver.exe"
driver = webdriver.Chrome(PATH)
driver.get('https://www.bewakoof.com/women-clothing')
page = driver.page_source
soup = BeautifulSoup(page,'lxml')
soup

p = []
count = 0
for main_file in soup.findAll('div',class_="productCardDetail"):
    if count != 10:
        name = main_file.h3.text
        price = main_file.b.text
        p.append([name,price])
        count += 1

i = []
count = 0
for img_url in soup.findAll('img',class_="productImgTag"):
    if count != 10:
        i.append(img_url['src'])
        count += 1
    
df = pd.DataFrame(p,columns=['Name','Price'])
df1 = pd.DataFrame(i)

df['Img_url'] = df1
df
```
