### Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You
### have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10
### jobs data.
### This task will be done in following steps:
1. First get the webpage https://www.naukri.com/
2. Enter “Data Analyst” in “Skill, Designations, Companies” field and enter “Bangalore” in “enter the
location” field.
3. Then click the search button.
4. Then scrape the data for the first 10 jobs results you get.
5. Finally create a dataframe of the scraped data.

```python
import selenium
from selenium import  webdriver
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
PATH= "C:\Program Files (x86)\chromedriver.exe"
driver = webdriver.Chrome(PATH)
```

```python
driver.get("https://www.naukri.com/")
```

```python
driver.find_element_by_class_name("suggestor-input ").send_keys("Data Analyst")
driver.find_element_by_xpath("/html/body/div/div[2]/div[3]/div/div[1]/div[3]/div/div/div/input").send_keys("Banglore")
driver.find_element_by_xpath("/html/body/div/div[2]/div[3]/div/div/div[6]").click()
```

```python
job_titles = []

title = driver.find_elements_by_xpath("//a [@class='title fw500 ellipsis']")
title

for i in title:
    t = i.text
    job_titles.append(t)
job_titles
```

```python
loc = []

location = driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']/span[1]")
location

for i in location:
    l = i.text
    loc.append(l)
loc
```

```python
company = []

company_name = driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']")
company_name

for i in company_name:
    c = i.text
    company.append(c)
company
```

```python
exp = []

exp_req = driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi experience']/span[1]")
exp_req

for i in exp_req:
    e = i.text
    exp.append(e)
exp
```

```python
jobs = pd.DataFrame({})
jobs['Title'] = job_titles
jobs['Location'] = loc
jobs['Company Name'] = company
jobs['Experience Required'] = exp

jobs[:10]
```

```python
PATH= "C:\Program Files (x86)\chromedriver.exe"
driver = webdriver.Chrome(PATH)
driver.get(" https://www.ambitionbox.com/")
```

```python
c = []
c_name = driver.find_elements_by_xpath("//div[@class='name']/a[1]")
for i in c_name:
    c.append(i.text)
c
```

```python
sal_rec = []
total_salr = driver.find_elements_by_xpath("//div[@class='name']/span[1]")
for i in total_salr:
    sal_rec.append(i.text.split(' ')[0].strip('('))
sal_rec
```

```python
avg = []
avg_sal = driver.find_elements_by_xpath("//div[@class='average-indicator-wrapper']/p[1]")
for i in avg_sal:
    avg.append(i.text)
avg[9:19]
```

```python
mis = []
min_sal = driver.find_elements_by_xpath("//div[@class='salary-values']/div[1]")
for i in min_sal:
    mis.append(i.text)
mis[9:19]
```

```python
mas = []
max_sal = driver.find_elements_by_xpath("//div[@class='salary-values']/div[2]")
for i in max_sal:
    mas.append(i.text)
mas[9:19]
```

```python
exp = []
exp_req = driver.find_elements_by_xpath("//div[@class='salaries one-line sbold-list-header']/span[1]")
for i in exp_req:
    exp.append(i.text)
exp
```

```python
DataScientist = pd.DataFrame({})
DataScientist['Company Name'] = c
DataScientist['Total Salary Record'] = sal_rec
DataScientist['Average Salary'] = avg[9:19]
DataScientist['Minimum Salary'] = mis[9:19]
DataScientist['Maximum Salary'] = mas[9:19]
DataScientist['Experience Required'] = exp

DataScientist
```
